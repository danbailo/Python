# Lista de Projetos

Esta lista contém cerca de 100 ideias de projetos para você experimentar em Python! Alguns deles são diretos e outros que fizemos antes (como uma seqüência de Fibonacci ou FizzBuzz).

Escolha um projeto simples que você acha que pode terminar em um dia para começar, depois escolha outro projeto que você acha que levará mais de uma semana!

Existem algumas soluções de exemplo nesta pasta, por isso sinta-se à vontade para explorar e lembrar:

**DIVERTA-SE!**

Números
---------

** Encontre o PI de "n" casas decimais. ** - Insira um número e faça com que o programa gere o PI até "n' casas decimais. Mantenha um limite para o quão longe o programa irá.

** Encontre "e" de "n" casas decimais. ** - Assim como o problema anterior, mas com "e" ao invés de PI. Digite um número e faça o programa gerar "e" até muitas casas decimais. Mantenha um limite para o quão longe o programa irá.

** Seqüência de Fibonacci ** - Digite um número e peça ao programa para gerar a sequência de Fibonacci para esse número ou para o número N.

** Fatoração Primo ** - Peça ao usuário que digite um número e encontre todos os números primos (se houver algum) e exiba-os.

** Next Prime Number ** - Peça ao programa para encontrar números primos até que o usuário opte por parar de perguntar pelo próximo.

** Encontre o Custo da Telha para Cobrir L x A ** - Calcule o custo total da telha necessária para cobrir uma planta baixa de largura e altura, usando um custo inserido pelo usuário.

** Programa de troco ** - O usuário insere um custo e depois a quantia de dinheiro dada. O programa calcular o troco (em reais e centavos).

** Binário para Decimal ** - Desenvolva um conversor para converter um número decimal em binário ou um número binário em seu equivalente decimal.

** Calculadora ** - Uma calculadora simples para fazer operadores básicos. Faça disso uma calculadora científica para maior complexidade.

** Conversor de Unidades (tempo, moeda, volume, massa e mais) ** - Converte várias unidades entre si. O usuário insere o tipo de unidade que está sendo inserida, o tipo de unidade que deseja converter e, em seguida, o valor. O programa fará então a conversão.

** Alarme ** - Um relógio simples, onde toca um som após o número X de minutos / segundos ou em um determinado momento.

** Distância entre duas cidades ** - Calcula a distância entre duas cidades e permite ao usuário especificar uma unidade de distância. Esse programa pode exigir a localização de coordenadas para as cidades, como latitude e longitude.

** Validador de cartão de crédito ** - Aceita um número de cartão de crédito de um fornecedor de cartão de crédito comum (Visa, MasterCard, American Express, Discoverer) e o valida para garantir que seja um número válido.

** Calculador de taxa ** - Solicita ao usuário que insira um custo e um imposto estadual ou nacional. Em seguida, retorna o imposto mais o custo total com imposto.

** Álgebra numérica complexa ** - Mostra adição, multiplicação, negação e inversão de números complexos em funções separadas. (Operações de subtração e divisão podem ser feitas com pares dessas operações.) Imprima os resultados para cada operação testada.

** Números Felizes ** - Um número feliz é definido pelo seguinte processo. Começando com qualquer número inteiro positivo, substitua o número pela soma dos quadrados de seus dígitos e repita o processo até que o número seja igual a 1 (onde permanecerá), ou faça um loop infinitamente em um ciclo que não inclua 1. Esses números para os quais este processo termina em 1 são números felizes, enquanto aqueles que não terminam em 1 são números infelizes. Mostre um exemplo de sua saída aqui. Encontre os primeiros 8 números felizes.

** Nomeando números ** - Mostra como soletrar um número em inglês. Você pode usar uma implementação preexistente ou criar sua própria, mas deve suportar entradas de no mínimo um milhão (ou o valor máximo do tipo de inteiro limitado padrão do seu idioma, se for menor). * Opcional: Suporte para entradas diferentes de inteiros positivos (como zero, inteiros negativos e números de ponto flutuante). *

** Cara ou Coroa ** - Escreva algum código que simule o lançamento de uma única moeda, por mais que o usuário decida. O código deve registrar os resultados e contar o número de caras ou coroas.

** Exponenciação Rápida ** - Peça ao usuário para inserir 2 inteiros aeb e calcule a ^ b (ou seja, pow(a, b)) em complexidade de tempo O(lg n).


Algoritmos Clássicos
-----------------

** Conjectura Collatz ** - Comece com um número *n> 1*. Encontre o número de etapas necessárias para chegar a 1 usando o seguinte processo: Se *n* for par, divida por 2. Se *n* for ímpar, multiplique por 3 e adicione 1.

** Classificação ** - Implemente dois tipos de algoritmos de ordenação: Merge sort e bubble sort (tente um Google).

** Problema de par mais próximo ** - O problema de par mais próximo de pontos ou de pares mais próximos é um problema de geometria computacional: dados *n* pontos no espaço métrico, encontre um par de pontos com a menor distância entre eles.

Peneira de Eratóstenes ** - A peneira de Eratóstenes é uma das formas mais eficientes de encontrar todos os primos menores (abaixo de 10 milhões ou mais).


Grafos
--------

** Grafos de links ** - Crie um programa que criará um grafo ou uma rede a partir de uma série de links.

** Caminho Euleriano ** - Crie um programa que terá como entrada um grafo e produza um caminho Euleriano ou um ciclo Euleriano, ou declare que isto não é possível. Um Caminho Euleriano inicia em um nó e percorre todas as arestas de um gráfico em cada nó e termina em outro nó. Um ciclo Euleriano é um caminho euleriano que inicia e termina no mesmo nó.

** Gráfico Conectado ** - Cria um programa que recebe um grafo como uma entrada e mostra se cada nó está conectado ou não.

** Algoritmo de Dijkstra ** - Crie um programa que encontre o caminho mais curto por meio de um grafo usando suas bordas.

** Árvore de Abrangência Mínima ** - Cria um programa que usa um grafo não direcionado conectado com pesos e gera a árvore geradora mínima do gráfico, ou seja, um subgrafo que é uma árvore, contém todos os vértices, e a soma de seus pesos é o menos possível.


Estruturas de dados
---------

** Índice invertido ** - Um [Índice Invertido](http://en.wikipedia.org/wiki/Inverted_index) é uma estrutura de dados usada para criar pesquisa de texto completo. Dado um conjunto de arquivos de texto, implemente um programa para criar um índice invertido. Também crie uma interface de usuário para fazer uma pesquisa usando esse índice invertido que retorna uma lista de arquivos que contêm o termo / termos da consulta. O índice de pesquisa pode estar na memória.


Texto
---------

** Fizz Buzz ** - Escreva um programa que imprima os números de 1 a 100. Mas para múltiplos de três imprima “Fizz” ao invés do número e para os múltiplos de cinco imprima “Buzz”. Para números que são múltiplos de três e cinco, imprima “FizzBuzz”.

** Reverse a String ** - Digite uma string e o programa irá invertê-la e imprimi-la.

** Pig Latin ** - Pig Latin é um jogo de alterações jogado no jogo da língua inglesa. Para criar a forma latina do porco de uma palavra inglesa, o som da consoante inicial é transposto para o final da palavra e é afixado (ex .: "banana" produziria anana-bay). Leia a Wikipedia para mais informações sobre regras.

** Contador de vogais ** - Digite uma string e o programa conta o número de vogais no texto. Para maior complexidade, informe a soma de cada vogal encontrada.

** Checa por palíndromes ** - Verifica se a string digitada pelo usuário é um palíndromo.

** Conte as palavras em uma string ** - Conta o número de palavras individuais em uma string. Para maior complexidade, leia essas sequências a partir de um arquivo de texto e gere um resumo.

** Editor de texto ** - Aplicativo de estilo do Bloco de notas que pode abrir, editar e salvar documentos de texto. * Opcional: adicione realce de sintaxe e outros recursos. *

** RSS Feed ** - Com um link para RSS / Atom Feed, obtenha todas as postagens e exiba-as.

** Livro de visitas / Diário ** - Um aplicativo simples que permite às pessoas adicionar comentários ou escrever entradas de diário. Pode permitir comentários ou não e carimbos de data e hora para todas as entradas. Também poderia ser transformado em uma caixa de mensagens. * Opcional: implante-o no Google App Engine ou Heroku ou qualquer outro PaaS (se possível, é claro). *

** Ferramenta de consulta Regex ** - Uma ferramenta que permite ao usuário inserir uma seqüência de texto e, em um controle separado, inserir um padrão de expressão regular. Ele executará a expressão regular no texto de origem e retornará quaisquer correspondências ou erros de sinalizador na expressão regular.



Networking
---------

** Programa de FTP ** - Um programa de transferência de arquivos que pode transferir arquivos de um servidor web remoto.

** Bandwidth Monitor ** - Um pequeno programa utilitário que controla quantos dados você carregou e baixou da rede durante a sessão online atual. Veja se você pode descobrir quais períodos do dia você usa mais e menos e gerar um relatório ou gráfico que o mostre.

** Port Scanner ** - Digite um endereço IP e um intervalo de portas em que o programa tentará encontrar portas abertas no computador em questão conectando-se a cada uma delas. Em qualquer conexão bem-sucedida, marque a porta como aberta.

** Mail Checker (POP3 / IMAP) ** - O usuário insere várias informações de conta, incluindo servidor da Web e IP, tipo de protocolo (POP3 ou IMAP), e o aplicativo verificará se há um intervalo de email.

** País da Pesquisa por IP ** - Digite um endereço IP e localize o país em que o IP está registrado. * Opcional: Encontre o IP automaticamente. *

** Whois Search Tool ** - Digite um IP ou endereço de host e procure por whois e retorne os resultados para você.

** Site Checker with Time Scheduling ** - Um aplicativo que tenta se conectar a um site ou servidor a cada tantos minutos ou um determinado momento e verificar se está funcionando. Se estiver desativado, ele será notificado por e-mail ou postando um aviso na tela.
 

Classes
---------

** Inventário de produto ** - Crie um aplicativo que gerencie um inventário de produtos. Crie uma classe de produto que tenha um preço, ID e quantidade à mão. Em seguida, crie uma classe *inventory* que rastreie vários produtos e possa somar o valor do inventário.

** Sistema de reserva de companhias aéreas / hotéis ** - Crie um sistema de reservas que reserve assentos de companhias aéreas ou quartos de hotel. Ele cobra várias taxas para seções específicas do avião ou hotel. Exemplo, primeira classe vai custar mais do que o treinador. Os quartos do hotel têm suítes na cobertura que custam mais. Acompanhe quando os quartos estarão disponíveis e poderão ser agendados.

** Gerente da Empresa ** - Criar uma hierarquia de classes - classe abstrata Empregado e subclasses do Gerente de Trabalho, Gerente e Executivo da HourlyEmployee, SalariedEmployee. O pagamento de cada um é calculado de forma diferente, pesquise um pouco sobre isso.
Depois de estabelecer uma hierarquia de funcionários, crie uma classe da empresa que permita gerenciar os funcionários. Você deve ser capaz de contratar, demitir e contratar funcionários.

** Gerente de Conta Bancária ** - Crie uma classe chamada Conta, que será uma classe abstrata para três outras classes chamadas CheckingAccount, SavingsAccount e BusinessAccount. Gerenciar créditos e débitos dessas contas através de um programa de estilo ATM.

** Agendador Paciente - Doutor ** - Crie uma classe de paciente e uma classe de médico. Tenha um objeto médico que possa lidar com vários pacientes e configurar um programa de agendamento, onde um médico só pode lidar com 16 pacientes durante um dia de trabalho de 8 horas.

** Recipe Creator and Manager ** - Crie uma receita com os ingredientes e coloque-os em um programa de gerenciamento de receita que os organiza em categorias como desertos, pratos principais ou ingredientes como frango, carne, sopas, tortas, etc.

** Galeria de Imagens ** - Crie uma classe abstrata de imagem e, em seguida, uma classe que herda dela para cada tipo de imagem. Coloque-os em um programa que os exiba em um formato de galeria para visualização.

** Área da forma e classes de perímetro ** - Crie uma classe abstrata chamada Shape e herde outras formas como diamante, retângulo, círculo, triângulo, etc. Depois, cada classe deve substituir a funcionalidade de área e perímetro para manipular cada tipo de forma.

** Floricultura ** - Crie uma aplicação de loja de flores que lida com objetos de flores e use esses objetos de flores em um objeto buquê que pode ser vendido. Acompanhe o número de objetos e quando você precisar pedir mais.


Threading
---------

** Criar uma barra de progresso para downloads ** - Crie uma barra de progresso para aplicativos que podem acompanhar o download em andamento. A barra de progresso estará em um thread separado e se comunicará com o thread principal usando delegates.

** Criador de miniaturas em massa ** - O processamento de imagens pode demorar um pouco para algumas transformações. Especialmente se a imagem for grande. Crie um programa de imagem que possa levar centenas de imagens e convertê-las em um tamanho especificado no encadeamento de plano de fundo enquanto você faz outras coisas. Para maior complexidade, tenha um thread administrando redimensionamento, tenha outra renomeação em massa de miniaturas etc.


Rede
---------

** Page Scraper ** - Crie um aplicativo que se conecta a um site e extrai todos os links, ou imagens, e os salva em uma lista. * Opcional: organize o conteúdo indexado e não permita duplicatas. Coloque os resultados em um arquivo de índice facilmente pesquisável. *

** Obter o horário de um relógio atômico ** - Este programa irá obter o verdadeiro tempo atômico a partir de um relógio atômico na Internet. Use qualquer um dos relógios atômicos retornados por uma simples pesquisa no Google.

** Buscar tempo atual ** - Obtenha o tempo atual para um determinado CEP / código postal. * Opcional: tente localizar o usuário automaticamente. *


** Criador CAPTCHA ** - Já viu essas imagens com letras de números quando você se inscrever para um serviço e, em seguida, pede-lhe para digitar o que vê? Ele impede que os robôs da web se inscrevam automaticamente e enviem spam. Tente criar um você mesmo para formulários on-line.


Arquivos
---------

** Quiz Maker ** - Faça um aplicativo que recebe várias perguntas de um arquivo, escolhido aleatoriamente e reúne um questionário para os alunos. Cada questionário pode ser diferente e, em seguida, lê uma chave para classificar os questionários.

** Classificar utilitário de arquivos Excel / CSV ** - Lê um arquivo de registros, classifica-os e, em seguida, grava-os novamente no arquivo. Permitir que o usuário escolha vários estilos de classificação e classificação com base em um campo específico.

** Gerenciador de arquivos Zips ** - O usuário insere vários arquivos de diretórios diferentes e o programa os coloca em um arquivo zip. * Opcional: Aplique a compactação real aos arquivos. Comece com o Algoritmo Huffman. *

** PDF Generator ** - Um aplicativo que pode ler em um arquivo de texto, arquivo html ou algum outro arquivo e gera um arquivo PDF a partir dele. Ótimo para um serviço baseado na web onde o usuário envia o arquivo e o programa retorna um PDF do arquivo. * Opcional: Implante no GAE ou no Heroku, se possível. *

** Mp3 Tagger ** - Modifique e adicione tags ID3v1 para arquivos MP3. Veja se você também pode adicionar a arte do álbum ao cabeçalho do arquivo MP3, além de outras tags ID3v2.



Banco de dados
---------

** SQL Query Analyzer ** - Um aplicativo utilitário no qual um usuário pode inserir uma consulta e executá-la em um banco de dados local e procurar maneiras de torná-lo mais eficiente.

** Ferramenta SQL Remota ** - Um utilitário que pode executar consultas em servidores remotos a partir do seu computador local pela Internet. Ele deve receber um host remoto, nome de usuário e senha, executar a consulta e retornar os resultados.

** Agendador de eventos e calendário ** - Faça um aplicativo que permite ao usuário inserir uma data e hora de um evento, notas do evento e, em seguida, agendar esses eventos em um calendário. O usuário pode então navegar no calendário ou pesquisar o calendário em busca de eventos específicos. * Opcional: permita que o aplicativo crie eventos de reincidência que ocorram todos os dias, semanas, meses, anos etc. *

** Budget Tracker ** - Escreva um aplicativo que acompanhe o orçamento de uma família. O usuário pode adicionar despesas, receitas e custos recorrentes para descobrir o quanto eles estão economizando ou perdendo ao longo de um período de tempo. * Opcional: permita que o usuário especifique um intervalo de datas e veja o fluxo líquido de entrada e saída do orçamento da casa para esse período de tempo. *

** TV Show Tracker ** - Tem um programa favorito que você não quer perder? Não tem um PVR ou quer ser capaz de encontrar o programa para depois PVR mais tarde? Faça um aplicativo que possa pesquisar vários sites do Guia de TV on-line, localize os programas / horários / canais e adicione-os a um aplicativo de banco de dados. O banco de dados / site pode enviar lembretes por e-mail de que um programa está prestes a começar e em qual canal ele estará.

** Travel Planner System ** - Crie um sistema que permita aos usuários montar seu próprio pequeno itinerário de viagem e acompanhe os arranjos aéreos / hoteleiros, pontos de interesse, orçamento e cronograma.


Gráficos e Multimídia
---------

** Stream Video Online ** - Tente criar seu próprio player de streaming de vídeo on-line.

** Mp3 Player ** - Um programa simples para reproduzir seus arquivos de música favoritos. Adicione recursos que você acha que estão faltando no seu player de música favorito.

** Aplicação de marca d'água ** - Tem algumas fotos que você quer protegidas por direitos autorais? Adicione o seu próprio logótipo ou texto de forma simples, para que ninguém possa simplesmente roubar os seus gráficos do seu site. Faça um programa que adicione essa marca d'água à imagem. * Opcional: use a segmentação para processar várias imagens simultaneamente. *

** Turtle Graphics ** - Este é um projeto comum onde você cria um piso de 20 x 20 quadrados. Usando vários comandos, você diz a uma tartaruga para desenhar uma linha no chão. Você se move para a frente, para a esquerda ou para a direita, levanta ou solta a caneta, etc. Faça uma busca online por "Turtle Graphics" para obter mais informações. * Opcional: permite que o programa leia na lista de comandos de um arquivo. *

** GIF Creator ** Um programa que reúne várias imagens (PNGs, JPGs, TIFFs) para criar um GIF suave que pode ser exportado. * Opcional: Faça o programa converter também pequenos arquivos de vídeo em GIFs. *
